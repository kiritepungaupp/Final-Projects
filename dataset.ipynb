```python
import whois
import datetime

def get_domain_age(domain: str) -> int:
    try:
        # Get WHOIS data for the domain
        w = whois.whois(domain)
        
        # Extract the creation date from the WHOIS data
        creation_date = w.creation_date
        if isinstance(creation_date, list):  # Sometimes it returns a list
            creation_date = creation_date[0]
        
        # Calculate the domain age (in years)
        if creation_date:
            age = (datetime.datetime.now() - creation_date).days 
            return age
        else:
            return 1  # Return 1 if creation date is not available
    except Exception as e:
        print(f"Error while fetching WHOIS data: {e}")
        return 1  # Return 1 if there's an error or domain is not found

```

```python
print(get_domain_age("https://www.instagram.com/"))
```
